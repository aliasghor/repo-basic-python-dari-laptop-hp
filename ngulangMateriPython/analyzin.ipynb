```python
import csv
from matplotlib import pyplot as plt
from collections import Counter
plt.style.use('dark_background')
```

```python
"""Read Csv File"""
with open('data.csv') as csv_file:
    csv_reader = csv.DictReader(csv_file)

    row = next(csv_reader)
    print(row)
```

```python
"""Printing Out The Key"""
with open('data.csv') as csv_file:
    csv_reader = csv.DictReader(csv_file)

    row = next(csv_reader)
    print(row['LanguagesWorkedWith'])
```

```python
"""Clean Up The Data By Splitting the (;)"""
with open('data.csv') as csv_file:
    csv_reader = csv.DictReader(csv_file)

    row = next(csv_reader)
    print(row['LanguagesWorkedWith'].split(';'))
```

```python
"""Printing Out And Looping The File"""
with open('data.csv') as csv_file:
    csv_reader = csv.DictReader(csv_file)

    languages_counter = Counter()

    for row in csv_reader:
        languages_counter.update(row['LanguagesWorkedWith'].split(';'))
print(languages_counter)
```

```python
"""Most Common 15 Languages"""
with open('data.csv') as csv_file:
    csv_reader = csv.DictReader(csv_file)

    languages_counter = Counter()

    for row in csv_reader:
        languages_counter.update(row['LanguagesWorkedWith'].split(';'))
print(languages_counter.most_common(15))
```

```python
"""Plotting The Data"""
with open('data.csv') as csv_file:
    csv_reader = csv.DictReader(csv_file)

    languages_counter = Counter()

    for row in csv_reader:
        languages_counter.update(row['LanguagesWorkedWith'].split(';'))

languages = []
popularity = []

for item in languages_counter.most_common(15):
    languages.append(item[0])
    popularity.append(item[1])

# print(languages)
# print(popularity)
languages.reverse()
popularity.reverse()
plt.barh(languages,popularity,color='r')

plt.title('Most Used Programming Languages')
plt.xlabel('Number Of People Who Use')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
```

```python
"""Reading Data With Pandas"""
data = pd.read_csv('data.csv')
ids = data['Responder_id']
lang_responses = data['LanguagesWorkedWith']

languages_counter = Counter()

for response in lang_responses:
    languages_counter.update(response.split(';'))

languages = []
popularity = []

for item in languages_counter.most_common(15):
    languages.append(item[0])
    popularity.append(item[1])

languages.reverse()
popularity.reverse()

plt.barh(languages,popularity,color='c')

plt.title('Most Used Programming Languages')
plt.xlabel('Number Of People Who Use')

plt.legend()
plt.tight_layout()
plt.show()
```
